```python
import pandas as pd
```

```python
df1=pd.read_parquet("yellow_tripdata_2023-01.parquet")
```

```python
df1.head()
```

## Q1. Downloading the data

```python
df1.shape
```

## Q2. Computing duration

```python
from datetime import datetime
```

```python
df1['tpep_pickup_datetime'] = pd.to_datetime(df1['tpep_pickup_datetime'], format="%Y-%m-%d %H:%M:%S")
df1['tpep_dropoff_datetime'] = pd.to_datetime(df1['tpep_dropoff_datetime'], format="%Y-%m-%d %H:%M:%S")
df1['difference'] = (df1['tpep_dropoff_datetime'] - df1['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
df1.head()
```

```python
df1['difference'].std()
```

## Q3. Dropping outliers

```python
df1['difference'].max()
```

```python
df1['difference'].describe()
```

```python
filtered_df = df1[(df1['difference'] >= 1) & (df1['difference'] <= 60)]

filtered_df.head()
```

```python
print(f'percentage left is: {(len(filtered_df)*100)/len(df1)}')
```

## Q4 One hot encoding

```python
df1.head()
```

```python
pu_do_df=filtered_df[['PULocationID','DOLocationID']]
pu_do_df.head()
```

```python
pu_do_df['PULocationID']=pu_do_df['PULocationID'].fillna(-1)
pu_do_df['DOLocationID']=pu_do_df['DOLocationID'].fillna(-1)
```

```python
pu_do_df['PULocationID']=pu_do_df['PULocationID'].astype(str)
pu_do_df['DOLocationID']=pu_do_df['DOLocationID'].astype(str)
```

```python
pu_do_dict = pu_do_df.to_dict(orient='records')
pu_do_dict
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
# v = DictVectorizer(sparse=False)
v = DictVectorizer()
```

```python
X = v.fit_transform(pu_do_dict)
X
```

```python
# X = v.transform(pu_do_dict)
```

```python
X.shape
```

## Q5. Training a model

```python
y_train = filtered_df['difference'].values
```

```python
y_train.shape
```

```python
X.shape
```

```python
len(v.feature_names_)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X, y_train)
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_train, y_pred, squared=False)
```
